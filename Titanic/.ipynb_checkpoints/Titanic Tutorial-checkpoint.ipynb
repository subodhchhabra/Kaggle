### Cleaning the data

```python
df['Gender'] = 4
df.head()
```

```python
df['Gender'] = df['Sex'].map( lambda x: x[0].upper())
df.head()
```

```python
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
df.head()
```
